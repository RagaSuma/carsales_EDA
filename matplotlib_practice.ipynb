```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,10)
```

```python
y= x**2
```

```python
plt.plot(x,y)
plt.show()
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Chart")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
plt.subplot(1,4,1)
plt.plot(x,y,'r')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Chart1")
plt.subplot(1,4,2)
plt.plot(y,x,'g')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Chart2")
plt.subplot(1,4,3)
plt.plot(x**2,y,'r')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Chart3")
plt.subplot(1,4,4)
plt.plot(y**2,x,'g')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Chart4")
```

```python
fig=plt.figure()
axis=fig.add_axes([0.1,0.1,0.9,0.9])
axis.set_xlabel('X Label')
axis.set_ylabel('Y Label')
axis.set_label('Chart')
axis.plot(x,y, 'r')
```

```python
fig=plt.figure()
axis1=fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis2=fig.add_axes([0.2,0.5,0.3,0.3])
axis2.plot(y,x,'r')
```

```python
fig, axis = plt.subplots(1,2)

for ax in axis:
    ax.plot(x,y)
```

```python
fig, axis = plt.subplots(1,2, figsize=(10,5))
axis[0].plot(x,y)
axis[1].plot(y,x,'r')
```

```python
fig.savefig('graph')
```

```python
fig=plt.figure()
axis=fig.add_axes([0.1,0.1,0.8,0.8])
axis.plot(x,x**2, label='Sqr root')
axis.plot(x,x**3, label='Cube root')
axis.legend()
```
