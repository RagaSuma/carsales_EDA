```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,6,10)
```

```python
x
```

```python
y = x **2
```

```python
#functional method
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Chart')
```

```python
#multiple plots together
#subplot

plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Chart1')

plt.subplot(2,2,2)
plt.plot(y,x,'b')
plt.xlabel('Y Label')
plt.ylabel('X Label')
plt.title('Chart2')

plt.subplot(2,2,3)
plt.plot(x**2,x,'b')
plt.xlabel('X sqr')
plt.ylabel('X Label')
plt.title('Chart3')

plt.subplot(2,2,4)
plt.plot(y,y ** 2,'b')
plt.xlabel('Y Label')
plt.ylabel('y sqr')
plt.title('Chart4')
```

```python
#object menthod

fig = plt.figure() #blank canvas
axis = fig.add_axes([0.1,0.1, 0.9, 0.9])
axis.plot(x,y)
axis.set_xlabel('X label')
axis.set_ylabel('Y Label')
axis.set_title('Title')
```

```python
#two charts

fig = plt.figure()

axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis1.set_xlabel('X label')
axis1.set_ylabel('Y Label')
axis1.set_title('Big Plot')


axis2 = fig.add_axes([0.2,0.5,0.3,0.3])
axis2.plot(y,x)
axis2.set_xlabel('Y label')
axis2.set_ylabel('X Label')
axis2.set_title('Small Plot')
#([%left, %bot, %width, %height])
```

```python
fig, axes = plt.subplots(1,2, figsize = (10,5))
axes[0].plot(x,y)
axes[1].plot(y,x)

fig.savefig('figurename.png')
```
