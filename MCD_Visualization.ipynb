```python
mcd_df2.plot(kind='hexbin',x='Carbohydrates',y='Energy',figsize=(15,7),colormap='Reds',gridsize=10)
plt.title('HEX plot showing showing the variation of energy served with each item')
```

```python
mcd_df2.groupby('Category')['Total_Fat'].mean().plot(kind='bar',figsize=(15,10),color='orange',grid=True)
mcd_df2.groupby('Category')['Saturated_Fat'].mean().plot(kind='bar',figsize=(15,7),color='green',grid=True)
mcd_df2.groupby('Category')['Trans_Fat'].mean().plot(kind='bar',figsize=(15,7),color='red',grid=True)
plt.xlabel('Category')
plt.ylabel('Fat Content')
plt.title('Stacked plot showing fat content per item')
plt.legend(labels=('Total_Fat','Saturated_Fat','Trans_Fat'))
```
